```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(np.random.uniform(-1,1,(10,10)),index = list(range(1,11)),columns=list(range(1,11)))
ext_data_corrs = df.corr()
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(ext_data_corrs, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)

plt.show()
```

```python
plot_data = pd.DataFrame(np.random.uniform(-1,1,(1000,3)),index = list(range(1,1001)),columns=['a','b','c'])

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
